# SQL

## FUNÇÕES NATIVAS SQL PARA TRABALHAR COM DATAS

#### YEAR, MONTH E DAY

As funções `YEAR`, `MONTH` E `DAY` são utilizadas para extrair o ano, o mês e o dia de datad ou datetimes.

#### NOW, CURDATE E CURTIME
As funções `NOW`, `CURDATE` e `CURTIME`, são usadas para retornar a a data e hora atual do sistema, a data atual do sistema e a hora atual do sistema, respectivamente.

#### DATEDIFF, DATE_ADD E DATE_SUB

- DATEDIFF: Restorna a diferença, em dias, entre duas datas;
- DATE_ADD: Adiciona um valor em dias, meses ou anos a uma data.
- DATE_SUB: Subtrai um valor em dias, meses ou anos a uma data.
