# SQL

## FUNÇÕES PARA CONVERSÃO DE DADOS

- **TRY_CONVERT**

`TRY_CONVERT` usa o valor passado a ele e tenta convertê-lo no data_type especificado. Se a conversão for bem-sucedida, `TRY_CONVERT` retornará o valor como o data_type especificado; se ocorrer um erro, será retornado nulo. Porém, se você solicitar uma conversão que não é permitida explicitamente, `TRY_CONVERT` falhará com um erro.

A cláusula `TRY_CONVERT` é específica do dialeto `T-SQL` e não está presente em outros dialetos SQL, como MySQL ou PostgreSQL, por exemplo.

Ao converter uma string vazia para um tipo numérico usando a função `TRY_CONVERT` no SQL Server, o resultado é zero. Isso ocorre porque o SQL Server interpreta uma string vazia como zero ao tentar convertê-la para um tipo numérico.

- **CONVERT**

A clásula convert é específica do dialeto T-SQL e sua sintaxe é `CONVERT(data_type(length), expression, style)`, _length_ somente se aplica aos tipos _CHAR_, _VARCHAR_, _NCHAR_, _NVARCHAR_, _BINARY_ e _VARBINARY_. _style_ é opcional e usado para formatação da saída dos dados.

Conversão de uma string, que representa uma data no formato dd/mm/yyyy, para o tipo DATETIME e a saída será formatada no padrão americano.

`CAST` também é uma função utilizada para converte um valor para outro tipo de dado, e poser ser usada no SQL Servre, MySQL, PostgreSQL e Oracle.
