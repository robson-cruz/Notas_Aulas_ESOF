#### LEFT

#### INSTR

A função `INSTR` em SQL é utilizada para encontrar a posição de uma substring dentro de uma string maior. Sua sintaxe geralmente é: `INSTR(string, substring)`

No exemplo a seguir, demonstraremos o uso combinado das funções `LEFT` e `INSTR` para extrair apenas o nome de usuário de endereços de email.

#### MID
