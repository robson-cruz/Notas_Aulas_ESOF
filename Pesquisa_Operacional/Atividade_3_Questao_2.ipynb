### Questão 2

<p style="text-align: justify; text-indent: 2rem">
    Em um processo de fabricação de computadores pessoais, são montados três tipos de computadores, sendo consumidos fios elétrico e fios de solda. Para o modelo A são utilizados 0,7m de fios elétricos e 0,3m de fios de solda, para o modelo B são utilizados 0,6m de fios elétricos e 0,35m de fios de solda, já para o modelo C, são utilizados 0,65m de fios elétricos e 0,25m de fios de solda. A quantidade disponível de fios elétricos no mês para essa produção é de 50m e de fios de solda de 35m. O lucro obtido pela venda de cada unidade dos computadores do modelo A é de R\$350,00, para o modelo B de R\$322,00 e para o modelo C de R\$355,00. As demandas mínimas de cada modelo são, 10 unidades de A, 7 unidades de B e 11 unidades de C.
</p>

<p style="text-align: justify; text-indent: 2rem">
    Utilizando a ferramenta Solver do Excel para encontrar o melhor mix de produção e considerando as variáveis de decisão valores inteiros, analise as afirmações a seguir.
</p>

<p style="text-align: justify; text-indent: 2rem">
    I) O relatório de resposta nos mostra que haverá sobra de 14,15m de fios de solda no mês, sendo a quantidade de fios elétricos limitante da capacidade do processo.
</p>

<p style="text-align: justify; text-indent: 2rem">
    II) Na solução ótima encontrada, verifica-se que são produzidas 56 unidades a mais do que o mínimo necessário para o computador do modelo C, isso pode ser observado no relatório de resposta.
</p>

<p style="text-align: justify; text-indent: 2rem">
    III) O lucro total encontrado para o período foi de R\$26.922,00 com a produção do mínimo do modelo A, um excedente de 4 unidades do modelo B e um excedente de 45 unidades do modelo C.
</p>

<p style="text-align: justify; text-indent: 2rem">
    É correto o que se afirma em:
</p>

Alternativ  2:

### Solução

Resumo dos dados

**Função Objetivo**

$ Max \quad Z =  350A + 322B + 355C$

**Restrições**

Fio Elétrico: $ 0,7A + 0,6B + 0,65C <= 50 $

Fio de Solda: $ 0,3A + 0,35B + 0,25C <= 35 $

Demanda A: $A \ge 10$

Demanda B: $B \ge 7$

Demanda C: $C \ge 11$

Não Negatividade: $A, B, C \ge 0$

A otimização mostrada na saída do Solver, indica que o lucro máximo é R\$ 26.922 e é obtido com a produção de 10 laptops do modelo A, 11 do modelo B e 56 do modelo C. Ao final do processo produtivo restarão 14,15 metros de fio de solda no mês.

Assim, as **afirmações corretas são I e III**, correspondentes a **alternativa 5**.
