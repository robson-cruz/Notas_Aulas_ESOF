### Questão 4

<p style="text-align: justify; text-indent: 2rem">
    (CESGRANRIO) Duas empresas concorrentes (ABC e XYZ) podem ter como estratégia atuar prioritariamente em um de três possíveis mercados (RJ, SP e MG). A depender da estratégia adotada por uma empresa, a outra tem uma significativa perda de mercado nacional, conforme indica o quadro a seguir, desenhado sob a ótica da empresa ABC.
</p>

<p style="text-align: justify; text-indent: 2rem">
    No caso das duas concorrentes (ABC e XYZ), adotando-se o método de eliminação sequencial de estratégias dominadas, básico no contexto da Teoria dos Jogos, conclui-se que a empresa ABC:
</p>

A) Ganhará 40% do mercado.

B) Ganhará 20% do mercado

C) Ganhará 10% do mercado


D) Continuará com a mesma fatia de mercado

E) Perderá 10% do mercado.


### Solução

Usando o critério Maximin e a estratégia Minimax:

O método de eliminação sequencial envolve a eliminação de estratégias que são sempre piores do que outra estratégia, independentemente do que o adversário faça.

Analisando a tabela, podemos ver que para a empresa ABC:

- Se a empresa XYZ escolher o mercado RJ, a melhor escolha para a ABC é o mercado RJ, pois +10% > +0%.

- Se a empresa XYZ escolher o mercado SP, a melhor escolha para a ABC é o mercado RJ, pois +20% > +0%.

- Se a empresa XYZ escolher o mercado MG, a melhor escolha para a ABC é o mercado SP, pois +50% > -10%.

Portanto, a estratégia MG é dominada para a empresa ABC e pode ser eliminada. Agora, temos:

| Empresa / Mercado |    |  XYZ |      |
|-------------------|----|------|------|
|                   |    | RJ   | SP   |
|       ABC         | RJ | +10% | +20% |
|                   | SP | +10% | +0%  |

Agora, para a empresa XYZ:

- Se a empresa ABC escolher o mercado RJ, a melhor escolha para a XYZ é o mercado SP, pois +20% > +10%.

- Se a empresa ABC escolher o mercado SP, a melhor escolha para a XYZ é o mercado RJ, pois +10% > +0%.

Portanto, a estratégia SP é dominada para a empresa XYZ e pode ser eliminada. Agora, temos:

| Empresa / Mercado |    |  XYZ |
|-------------------|----|------|
|                   |    | RJ   |
|       ABC         | RJ | +10% |
|                   | SP | +10% |

Neste ponto, não há mais estratégias dominadas para eliminar. Portanto, a estratégia ótima para a empresa ABC é indiferente entre RJ e SP, e ambas resultam em um ganho de +10% no mercado.

Portanto, a resposta correta é a alternativa **C) Ganhará 10% do mercado.**

Utilizando o critério Maximin e estratégia Minmax:

| Empresa / Mercado |    |      |  XYZ |      |     |
|-------------------|----|------|------|------|-----|
|                   |    | RJ   | SP   | MG   | Min Linahs |
|       ABC         | RJ | +10% | +20% | +40% | +10% |
|                   | SP | +10% | +0%  | +50% | +0% |
|                   | MG | +0% | +10%  | -10% | -10% |
|                   |Máx Colunas | +10% | +20% | +40%| 

i) Determinar, em cada linha, a recompensa mínima

ii) Determinar, em cada coluna, a recompensa máxima.

iii) determinar o ponto de sela, que corresponde ao ponto que seja ao mesmo tempo o máximo e commun em coluna e em linha.
