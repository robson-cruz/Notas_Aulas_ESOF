### Solução

Resumo dos dados

**Função Objetivo**

Sejam: $x_{1} = Taças$; $x_{2} = Jarros$ e $x_{3} = Castiçais$

$ Max \quad Z =  15x_{1} + 25x_{2} + 40x_{3}$

**Restrições**

Tempo de Processo: $ 5x_{1} + 7x_{2} + 15x_{3} <= 900 $

Tempo de Embalagem: $ x_{1} + 2x_{1} + 10x_{1} <= 420 $

Quantidade de Vidro para $x_{1}$: $x_{1} \ge 1,5$

Quantidade de Vidro para $x_{2}$: $x_{2} \ge 3$

Quantidade de Vidro para $x_{3}$: $x_{3} \ge 2$

Não Negatividade: $x_{1}, x_{2}, x_{3} \ge 0$

A otimização mostrada na saída do Solver, indica que o lucro máximo é R\$ 3.200 e é obtido com a produção de 128 jarros, sendo necessários 384 kg de vidro, 256 minutos para embalagem e 896 minutos para o processo. Ao final do processo produtivo restarão 346 kg de vidro.

Assim, as **afirmações corretas são I e II**, correspondentes a **alternativa 2**.
