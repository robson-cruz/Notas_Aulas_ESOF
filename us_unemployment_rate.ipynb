```python
import pandas as pd
```

```python
unemployment = pd.read_csv('./us_unemployment_rate.txt')
```

```python
display(unemployment.head())
```

```python
unemployment_clean = unemployment.melt(id_vars=['Year'], var_name='month', value_name='unemployment_rate')
display(unemployment_clean.head())
display(unemployment_clean.head())
```

```python
unemployment_clean['month'] = [str(i) for i in unemployment_clean['month']]
unemployment_clean['Year']= [str(i) for i in unemployment_clean['Year']]

unemployment_clean['date'] = pd.to_datetime(unemployment_clean['month'] + '-' + unemployment_clean['Year'])
display(unemployment_clean)
```

```python
import matplotlib.pyplot as plt

plt.style.use('ggplot')
#plt.figure(figsize=(18, 10))
unemployment_clean.plot(x='date', y='unemployment_rate', legend=False)
plt.xlabel('')
plt.show()
```
